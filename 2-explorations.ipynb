Welcome to the second notebook!  The purpose of this notebook is exploratory data analysis: to get a feel, preferably visually, for which features do best at predicting default.

We start with the same train/test split as we always do on this data.

```python
import pandas as pd
df = pd.read_csv("../Data/default.csv", skiprows=1, index_col = "ID")

from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df.copy(),
                                        shuffle = True,
                                        random_state = 167,
                                        test_size = .2,
                                        stratify = df["default payment next month"])
```

First we'll explore the continuous variables.  For each, we want to compare the histogram of distribution of values for the defaulters versus the nondefaulters, to see if we visually notice a difference.  However, there are few issues:
- The numbers of defaulters and nondefaulters are so different that it is difficult to compare histograms without rescaling.  To deal with this, we plot the histogram of a smaller set of nondefaulters, with size equal to the full number of defaulters.
- The width of the histogram bars are automatically determined from the min and max of the data values, so they can change a lot depending on the inclusion of a few outliers.  To deal with this, we exclude outliers first before plotting.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
def compare_hist(df_in, col_name):
    plt.figure(figsize=(14,4))
    
    df_temp = df_in[[col_name, "default payment next month"]].copy()
    df_temp.rename(columns={"default payment next month":"defq"}, inplace=True)
    
    # take the interval containing all but the lowest and highest 10% of the data,
    # scale it by 3, and drop all data outside that range as outliers
    percent = 10
    upper,lower = np.percentile(df_temp[c],[100-percent,percent])
    main_range = upper-lower
 
    mymax = upper+(1.5*main_range)
    mymin = lower-(1.5*main_range)
    
    df_temp = df_temp[(df_temp[c] > mymin) & (df_temp[c] < mymax)]
    
    numdefs = sum(df_temp["defq"]==1)
    
    ## Get observations for this column where y = 0, a number of them equal to the number of defaulters
    plt.hist(df_temp[c][df_temp["defq"]==0][:numdefs], histtype="step", color='b', label="no default", bins=50)
    
    ## Get all observations for this column where y = 1
    plt.hist(df_temp[c][df_temp["defq"]==1], histtype="step", color='r', label="yes default", bins=50)
    
    plt.legend(fontsize=14)
    plt.title(c,fontsize=16)
    plt.show()
```

```python
for c in ["LIMIT_BAL", "AGE", "BILL_AMT1", "PAY_AMT1"]:
    compare_hist(df_train, c)
```

It looks like defaulters are more likely to have a low credit limit, and more likely to have paid close to 0 in the most recent month.  There might be some trends related to age, or perhaps to the amount of the most recent bill, but they're not obvious.

Let's make some new continuous variables by combining various existing continuous variables, and see if they show any trends.  Just to have it all in one place, the following function defines all the variables we could think of, including some that don't turn out to be very useful at all.

```python
def expand_columns(df_original):
    df_new = df_original.copy()
    df_new.rename(columns={"PAY_0":"PAY_1"}, inplace=True)
    
    # clean up the payment lateness so each month is either paid in full, paid on time carrying balance, or late
    for i in range(1, 7):
        df_new["CLEANPAY_"+str(i)] = df_new["PAY_"+str(i)].copy()
        df_new.loc[df_new["CLEANPAY_"+str(i)] > 0, "CLEANPAY_"+str(i)] = 1
        df_new.loc[df_new["CLEANPAY_"+str(i)] < 0, "CLEANPAY_"+str(i)] = -1

    # clean up demographic info    
    df_new["CLEANSEX"] = df_new["SEX"].copy()
    df_new.loc[(df_new.CLEANSEX != 1) & 
               (df_new.CLEANSEX != 2), "CLEANSEX"] = 2 # if not male or female, assign female
    df_new["CLEANMARRIAGE"] = df_new["MARRIAGE"].copy()
    df_new.loc[(df_new.CLEANMARRIAGE != 1) & 
               (df_new.CLEANMARRIAGE != 2), "CLEANMARRIAGE"] = 1.5 # assign other marital status to intermediate
    # make one-hot encoding for both marriage and cleanmarriage
    df_new = pd.concat([df_new, 
                        pd.get_dummies(df_new["MARRIAGE"], prefix="MARRIAGE"), 
                        pd.get_dummies(df_new["CLEANMARRIAGE"], prefix="CLEANMARRIAGE")], 
                       axis=1)
    
    # break EDUCATION into three binary variables
    df_new["GRAD"] = 1*(df_original.EDUCATION == 1)
    df_new["UNIV"] = 1*((df_original.EDUCATION == 1) | (df_original.EDUCATION == 1))
    df_new["EDU_OTHER"] = 1*((df_original.EDUCATION != 1) & 
                             (df_original.EDUCATION != 2) & 
                             (df_original.EDUCATION != 3))
#     df_new.drop(columns=["EDUCATION"], inplace=True)
    df_new["CLEANEDU"] = df_new["EDUCATION"].copy()
    df_new.loc[(df_new.EDUCATION != 1) & 
               (df_new.EDUCATION != 2) & 
               (df_new.EDUCATION != 3), "CLEANEDU"] = 4
    df_new = pd.concat([df_new, 
                        pd.get_dummies(df_new["EDUCATION"], prefix="EDU"), 
                        pd.get_dummies(df_new["CLEANEDU"], prefix="CLEANEDU")], 
                       axis=1)

    for i in range(1, 6):
    # record the portion of last bill that was not paid down each month
        df_new["DEBT_AMT"+str(i)] = df_new["BILL_AMT"+str(i+1)] - df_new["PAY_AMT"+str(i)]    
    for i in range(1, 6):
    # record the amounts of new charges for each bill
        df_new["CHARGE_AMT"+str(i)] = (df_new["BILL_AMT"+str(i)] - df_new["DEBT_AMT"+str(i)])
    for i in range(1, 5):
        # are they paying down their debt or increasing it?
        df_new["DEBT_DIFF"+str(i)] = df_new["DEBT_AMT"+str(i)] - df_new["DEBT_AMT"+str(i+1)]
    for i in range(1, 7):
        df_new["BILL_FRAC"+str(i)] = df_new["BILL_AMT"+str(i)]/df_new["LIMIT_BAL"]
    for i in range(1, 6):
    # record the fraction of last bill that was paid
        df_new["PAY_FRAC"+str(i)] = df_new["PAY_AMT"+str(i)]/df_new["BILL_AMT"+str(i+1)]
        df_new.loc[df_new["BILL_AMT"+str(i+1)] <= 0, "PAY_FRAC"+str(i)] = 1 # if nothing is billed, they paid in full
    for i in range(1, 6):
    # record the fraction of the limit balance that is the debt
        df_new["DEBT_FRAC"+str(i)] = df_new["DEBT_AMT"+str(i)] / df_new["LIMIT_BAL"]


#     df_new.drop(columns=["BILL_AMT"+str(i) for i in range(1, 7)], inplace=True)

    return df_new
```

```python
for c in ["DEBT_AMT1", "CHARGE_AMT1", "DEBT_DIFF1", "BILL_FRAC1", "PAY_FRAC1", "DEBT_FRAC1"]:
    compare_hist(expand_columns(df_train), c)
```

Surprisingly, among both defaulters and nondefaulters there seem to be many customers who carry close to 0 balance, have close to 0 new charges, have close to 0 change in the balance they carry, have a bill that's close to a 0 fraction of their credit limit, and/or pay close to the full bill.

Let's see what these histograms of continuous variables look like if we restrict our attention to just the customers who pay on time each month with a balance, or just the customers who pay on time each month without a balance.

```python
df_pre = expand_columns(df_train)

#customers who pay on time each month, with a balance carried forward
df_zeros = df_pre[(df_pre["PAY_1"]== 0) & 
                  (df_pre["PAY_2"]== 0) & 
                  (df_pre["PAY_3"]== 0) & 
                  (df_pre["PAY_4"]== 0) & 
                  (df_pre["PAY_5"]== 0) & 
                  (df_pre["PAY_6"]== 0)]

#customers who pay whole bill on time each month, so without a balance carried forward.
df_fulls = df_pre[(df_pre["PAY_1"] < 0) & 
                  (df_pre["PAY_2"] < 0) & 
                  (df_pre["PAY_3"] < 0) & 
                  (df_pre["PAY_4"] < 0) & 
                  (df_pre["PAY_5"] < 0) & 
                  (df_pre["PAY_6"] < 0)]
print(df_zeros["default payment next month"].value_counts(), 
      "\n\n", 
      df_fulls["default payment next month"].value_counts())
```

```python
for c in ["LIMIT_BAL", "AGE", "BILL_AMT1", "PAY_AMT1", "DEBT_AMT1", "CHARGE_AMT1", "DEBT_DIFF1", "BILL_FRAC1", "PAY_FRAC1", "DEBT_FRAC1"]:
    compare_hist(df_zeros, c)
```

```python
for c in ["LIMIT_BAL", "AGE", "BILL_AMT1", "PAY_AMT1", "DEBT_AMT1", "CHARGE_AMT1", "DEBT_DIFF1", "BILL_FRAC1", "PAY_FRAC1", "DEBT_FRAC1"]:
    compare_hist(df_fulls, c)
```

Compared to peers with the same payment lateness history, defaulters seem poorer overall: smaller bills, smaller payments, and smaller new charges.  Among those carrying a balance, defaulters have a smaller credit limit, and their bills and their debt are a larger proportion of their credit limit.  Among those not carrying a balance, defaulters don't have an obviously smaller credit limit, so their bills are a smaller fraction of their credit limit, because they are most likely to be close to zero.

Next let's look at the discrete variables, and make tables to show whether default is more or less common among those with a particular value.

```python
def proportions(df_in, cols):
    defq = "default payment next month"
    base_ratio = sum(df_in[defq] == 1)/len(df_in)
    all_counts = df_in[cols].value_counts().to_frame().rename(columns={0:"all"})
    def_counts = df_in[df_in[defq] == 1][cols].value_counts().to_frame().rename(columns={0:"def"})
    both_counts = all_counts.join(def_counts) 
    
    #multiplier: default rate in category(/ies) 'cols', relative to overall default rate in data
    both_counts["multiplier"] = both_counts["def"]/both_counts["all"]/base_ratio
    return both_counts#.sort_values("multiplier", ascending=False)
```

```python
proportions(df_pre, ["PAY_1"])
```

This table says that, for instance, of the 2999 who were one month late on the most recent payment, 1025 of them defaulted, which is 1.545 times the rate of default overall.  This single column is the best predictor of default.
